```python
def setimg(img, start, lenn=32):
    for i in range(32):
        for j in range(lenn):
            pic[start * 32 + j, i, :] = img[j, i, :]
def sel(img, i, j):
    return img[32*(i-1):32*i,32*(j-1):32*j]
```

```python
#gen mif core
core = cv2.imread('result.jpg')
mystr = ""
lenx, leny, _ = core.shape
nx = lenx/32
for i in range(nx):
    for j in range(4):
        for ii in range(32):
            for jj in range(32):
                mystr += str(hex(core[i*32+ii,j*32+jj,2]/ 16))[2:]
                mystr += str(hex(core[i*32+ii,j*32+jj,1]/ 16))[2:]
                mystr += str(hex(core[i*32+ii,j*32+jj,0]/ 16))[2:]
                mystr += " "
            mystr += "\n"
with open('frame.mif',"w") as f :
    f.write(mystr)
```

```python
#gen mif core
core = cv2.imread('p.jpg')
mystr = ""
lenx, leny, _ = core.shape
for i in range(lenx):
    for j in range(leny):
        mystr += str(hex(core[i,j,0]/ 16))[2:]
        mystr += str(hex(core[i,j,1]/ 16))[2:]
        mystr += str(hex(core[i,j,2]/ 16))[2:]
        mystr += " "
    mystr += "\n"
with open('frame.mif',"w") as f :
    f.write(mystr)
```

```python
#gen ip core
core = cv2.imread('result.jpg')
lenx, leny, _ = core.shape
mystr = "memory_initialization_radix=16;\n"
mystr += "memory_initialization_vector=\n"
for i in range(lenx):
    for j in range(leny / 4):
        for k in range(4):
            mystr += str(hex(core[i,4*j+k,0]/16))[2:]
            mystr += str(hex(core[i,4*j+k,1]/ 16))[2:]
            mystr += str(hex(core[i,4*j+k,2]/ 16))[2:]
            mystr += ","
        mystr += "\n"
with open('frame.coe',"w") as f :
    f.write(mystr)
```

```python
mapp = np.array(pd.read_excel('map.xlsx')).astype(np.uint8)
mystr = ""
lenx, leny = mapp.shape
now = 0;
nx = lenx / 11
for k in range(nx):
    for j in range(11):
        for i in range(11):
            mystr += "map[%d]=8'd%d;" % (k * 15 * 20 + j * 16 + i,mapp[k*11 + i,j])
            now = now + 1
        mystr += "\n"
    mystr += "\n"
with open('map.mif',"w") as f :
    f.write(mystr)
```
